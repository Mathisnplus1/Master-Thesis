```python
import matplotlib.pyplot as plt
import numpy as np

def draw_circle(ax, center, radius=0.05, color='w', edgecolor='k'):
    circle = plt.Circle(center, radius, fc=color, ec=edgecolor, lw=2)
    ax.add_patch(circle)

def draw_line(ax, start, end, color='k'):
    line = plt.Line2D([start[0], end[0]], [start[1], end[1]], color=color)
    ax.add_line(line)

# Create a figure and axis
fig, ax = plt.subplots()
ax.set_aspect('equal')

# Parameters for layers
input_layer_size = 3
hidden_layer_size = 4
output_layer_size = 2

# Node positions
input_layer_nodes = np.linspace(0.8, -0.8, input_layer_size)
hidden_layer_nodes = np.linspace(0.8, -0.8, hidden_layer_size)
output_layer_nodes = np.linspace(0.3, -0.3, output_layer_size)

# Draw input layer
for i, y in enumerate(input_layer_nodes):
    draw_circle(ax, (-0.5, y))

# Draw hidden layer
for i, y in enumerate(hidden_layer_nodes):
    draw_circle(ax, (0, y))

# Draw output layer
for i, y in enumerate(output_layer_nodes):
    draw_circle(ax, (0.5, y))

# Draw connections
for i in input_layer_nodes:
    for j in hidden_layer_nodes:
        draw_line(ax, (-0.5, i), (0, j))

for i in hidden_layer_nodes:
    for j in output_layer_nodes:
        draw_line(ax, (0, i), (0.5, j))

# Set plot limits
ax.set_xlim(-1, 1)
ax.set_ylim(-1, 1)

# Remove axes
ax.axis('off')

# Display the plot
plt.show()
```
