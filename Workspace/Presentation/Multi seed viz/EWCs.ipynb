```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from datetime import datetime


mpl.rcParams['axes.labelsize'] = 18
mpl.rcParams['xtick.labelsize'] = 18
mpl.rcParams['xtick.major.size'] = 5
mpl.rcParams['xtick.major.width'] = 2
mpl.rcParams['ytick.major.size'] = 5
mpl.rcParams['ytick.major.width'] = 2
mpl.rcParams['ytick.labelsize'] = 18
mpl.rcParams['legend.fontsize'] = 14
```

```python
EWC_nous = np.mean([89.16,92.18,87.86,85.67,80.27,81.17,60.56,76.75,60.06,88.7]) # Results/EWC_4
EWC_nous = 83
```

```python
EWC0 = 97 # https://arxiv.org/abs/1612.00796
EWC1 = 84 # https://openreview.net/forum?id=BkQqq0gRb
EWC2 = 93 # https://arxiv.org/pdf/1708.02072
EWC3 = 95.05 # https://arxiv.org/pdf/2109.10021
EWC4 = 89.54 # https://proceedings.neurips.cc/paper_files/paper/2021/file/901797aebf0b23ecbab534d61ad33bb1-Paper.pdf
EWC5 = 82.1 # https://www.sciencedirect.com/science/article/pii/S0888613X23002360
EWC6 = 88.2 #https://arxiv.org/pdf/2003.09553
EWC7 = 94.31 #https://arxiv.org/pdf/1904.07734
```

```python
EWCs = [EWC1, EWC2, EWC3, EWC4, EWC5, EWC6, EWC7]
```

```python
# Create the plot
plt.figure(figsize=(4.5, 6))  # Set the figure size to be wide and short

# Plot data points as scatter along the x-axis
plt.scatter(0, EWC0, c="red", label="Original paper", s=60)
plt.scatter(0, EWC_nous, c="blue", label="Our reproduction", s=60)
plt.scatter(np.zeros_like(EWCs), np.array(EWCs), c="black", label="Other reproductions", s=60)  # y values are all zeros to align them horizontally

# Customize the plot
plt.xticks([])
plt.xlim([-0.1, 0.5])
#plt.xlabel('Value')
plt.ylabel('Test accuracy')

plt.legend(loc="upper right")
plt.tight_layout()

# Save
if True : #savefig and folder is not None :
    current_time = datetime.now().strftime("%Y-%m-%d_%H-%M-%S")
    plt.savefig(f"EWCs_{current_time}.png")

plt.show()
```
